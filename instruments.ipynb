```python
import requests
import pandas as pd
import defs
```

```python
session = requests.Session()
```

```python
instrument = "EUR_USD"
count = 10
granularity = "H1"
defs.OANDA_URL
```

```python
url = f"{defs.OANDA_URL}/accounts/{defs.ACCOUNT_ID}/instruments"
url
```

```python
params = dict(
    count = count,
    granularity = granularity,
    price = "MBA"
)
```

```python
response = session.get(url, params=None, headers=defs.SECURE_HEADER)
```

```python
response.json()
```

```python

# Convertir la respuesta JSON a DataFrame
data = response.json()
instruments_df = pd.DataFrame(data['instruments'])

# Mostrar el DataFrame
instruments_df

```
